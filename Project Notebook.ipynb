### Factor 1: Waiting Day

```python
import numpy as np
# To use np.timedelta64 : 
# The arguments for timedelta64 are:
# a number to represent the number of units, 
# and a date/time unit such as (D)ay, (M)onth, (Y)ear, (h)ours, (m)inutes, or (s)econds.
```

```python
# Calculating waiting days by subtracting Scheduled_date from Appointment_date
# Waiting_days = waiting period (in days) experienced by a patient between scheduled date & appointment date.
app_df["Waiting_days"] = (app_df["Appointment_date"].dt.date - app_df["Scheduled_date"].dt.date)/np.timedelta64(1,'D')
app_df["Waiting_days"].head()
```

```python
# Create a new column that categorizes waiting days by months
app_df['WaitingMonth'] = (app_df['Waiting_days'] // 30) + 1
```

```python
waiting_table = app_df.groupby(['WaitingMonth', 'No_show']).size().unstack(fill_value=0)
display(waiting_table)
```

### Visual Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
# Step 1: Define bins and labels
bins = [-1, 0, 7, 14, 21, 28, 56, 84, app_df['Waiting_days'].max()]
labels = [
    "same day", "1 week", "2 week", "3 week", "4 week",
    "two months", "three months", "more than three months"
]
# Step 2: Create a new categorical column
app_df["WaitingCategory"] = pd.cut(app_df["Waiting_days"], bins=bins, labels=labels, right=True, ordered=True)
# Step 3: Group and count Showed up vs Missed Appointment per category
grouped = app_df.groupby(["WaitingCategory", "No_show"]).size().unstack(fill_value=0)
# Step 4: Plot
grouped.plot(
    kind="bar",
    figsize=(10, 6),
    color={"Showed up": "green", "Missed Appointment": "red"},
    width=0.6
)
plt.title("Appointment Status by Waiting Time Category")
plt.xlabel("Waiting Period Category")
plt.ylabel("Number of Patients")
plt.xticks(rotation=45)
plt.legend(title="Appointment Status")
plt.tight_layout()
plt.show()
```

### Conclusion:
- The chart shows that most appointments were scheduled with short waiting times, especially on the same day and within 1 to 2 weeks. As waiting time increased, the total number of patients in each category decreased.
- While the number of missed appointments (in red) increases slightly in categories like “2 weeks” or “two months”, the overall proportion of no-shows compared to those who showed up remains relatively stable across all waiting time categories.
- This suggests that longer waiting times alone may not strongly increase the chance of missing an appointment.
